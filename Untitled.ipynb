```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
def f(x,c,b):
    """
    Define the function to fit with curve_fit.
    
    Parameters
    ----------
    x: array
        Change in displacement values
    c: int
        Speed of light variable
    b: int
        Y-intercept
    """
    #Define the distance measured in lab
    r=int(4895.85) #mm
    D=int(14861.54) #mm
    
    m = c/(4*np.pi*r*D) #slope from working equation 
    return  m*x+b #linear fit

x =np.array([1e-30,0.155,0.17,0.34,0.36,0.15,0.21,0.23,0.16,0.17,0.26,0.14,0.12,0.08]) #change in displacement (mm)
y =np.array([1e-30,146.5,46.5,77,66.4,76.9,77,82,46.3,61.7,34.3,70.6,42.8,35]) #Change in frequency (Hz)
uncertainty = np.array([1e-30,0.5,1,1,0.6,1,1.5,2,2,3.6,4,5.5,5,6]) #Frequency uncertainties

p0 = 3e11,0 #initial guess values 

popt, pcov = curve_fit(f, x, y, p0) #unweighted fit
popt2, pcov2 = curve_fit(f, x, y, p0,sigma=uncertainty, absolute_sigma=True) #weighted fit 

print('Unweighted fit parameters:', popt)
print('Covariance matrix:', np.sqrt(np.diag(pcov))) #std of c and b

print('Weighted fit parameters:', popt2)
print('Covariance matrix:', np.sqrt(np.diag(pcov2))) #std of c and b
```

```python
plt.figure(figsize=(20,10))
plt.plot(x,f(x,*popt2), label ='Weighted Linear Least Squares', color='red')
plt.plot(x,f(x,*popt), label ='Unweighted Linear Least Squares', color='blue')
plt.errorbar(x,y,yerr=uncertainty,fmt='o',label ='Measured Data', color='black')
plt.legend(fontsize=20)
plt.ylabel("$\Delta$ Frequency (Hz)", fontsize =20)
plt.xlabel("$\Delta$ Displacement (mm)", fontsize =20)
plt.title("Unweighted vs. Weighted Linear Least Square Fits", fontsize =30)
plt.text(0,130,"Weighted:", fontsize=20, color='red')
plt.text(0,120,'c = 2.98146119e+11 mm/s ', fontsize=15, color='red')
plt.text(0,110,'b = -3.25150889e-12 Hz ', fontsize=15, color='red')
plt.text(0,100, "Unweighted:", fontsize=20, color='blue')
plt.text(0,90,'c = 1.16837721e+11 mm/s ', fontsize=15, color='blue')
plt.text(0,80,'b = 3.84084161e+01 Hz ', fontsize=15, color='blue')
plt.savefig('speedoflightlab.jpeg')
```
